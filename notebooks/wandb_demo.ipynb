```python
%load_ext autoreload
%autoreload 2

import sys
import os


module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

import wandb
from src.utils.data_loader import load_and_preprocess_data

wandb.init(project="DA6401_Assignment_1-src", job_type="data_exploration")
```

```python
import numpy as np

(X_train, y_train), _, _ = load_and_preprocess_data(dataset_name="mnist")


class_names = [str(i) for i in range(10)]


table = wandb.Table(columns=["Class", "Image_1", "Image_2", "Image_3", "Image_4", "Image_5"])

print("Generating image table...")
for class_idx in range(10):
    
    indices = np.where(y_train== class_idx)[0][:5]
    
    
    row_images = []
    for idx in indices:
        
        img_pixels = X_train[idx].reshape(28, 28)
        row_images.append(wandb.Image(img_pixels))
    
    table.add_data(class_names[class_idx], *row_images)


wandb.log({"mnist_exploration": table})
wandb.finish()
print("Table successfully logged to W&B!")
```
